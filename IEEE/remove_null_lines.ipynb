```python
for y in range(1905, 1946):
    fn = str(y)+".txt"
    print(fn)
    lines = []
    nullcount = 0
    with open(fn, "r", encoding="utf8") as file:
        lines = file.readlines()
        
    for i in range(len(lines)):
        if lines[i] == 'null\n':
            nullcount += 1
            lines[i] = ""

    print("found " + str(nullcount) + " nulls.")
    
    with open(fn, "w", encoding="utf8") as file:
        file.writelines(lines)
```
